```python
import kagglehub
import pandas as pd
import os
import chess
# import gym
# import gymnasium as gym
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

dataset_path = os.path.join(kagglehub.dataset_download("datasnaek/chess"), "games.csv")
data_csv = pd.read_csv(dataset_path)
```

```python
data_csv
```

```python
def moves_to_fen_list(moves_str):
    board = chess.Board()
    fen_list = [board.fen()]  # Start with initial position
    
    # Split moves string into individual moves
    moves = moves_str.strip('...').split()
    
    # Apply each move and store resulting FEN
    for move in moves:
        try:
            board.push_san(move)
            fen_list.append(board.fen())
        except:
            break  # Stop if invalid move encountered
            
    return fen_list

# Apply the function to the dataframe
data_csv['fen_positions'] = data_csv['moves'].apply(moves_to_fen_list)

# Example output for the first row
print("First game FEN positions:")
for i, fen in enumerate(data_csv['fen_positions'].iloc[0]):
    print(f"Move {i}: {fen}")
```

```python
# Flatten all FEN positions into a single list
all_fen_positions = [fen 
                    for game_fens in data_csv['fen_positions'] 
                    for fen in game_fens]

print(f"Total number of positions: {len(all_fen_positions)}")
print("\nFirst 5 positions:")
for i, fen in enumerate(all_fen_positions[:5]):
    print(f"{i}: {fen}")
```

```python
# Set sampling factor (e.g., 0.1 for 10% of positions)
sampling_factor = 0.02

# Randomly sample positions
sampled_positions = np.random.choice(
    all_fen_positions, 
    size=int(len(all_fen_positions) * sampling_factor),
    replace=False
)

print(f"Number of sampled positions: {len(sampled_positions)}")
print("\nFirst 5 sampled positions:")
for i, fen in enumerate(sampled_positions[:5]):
    print(f"{i}: {fen}")
```

```python
# Save sampled positions to a text file
output_file = 'sampled_positions.txt'
with open(output_file, 'w') as f:
    for fen in sampled_positions:
        f.write(f"{fen}\n")

print(f"Saved {len(sampled_positions)} positions to {output_file}")
```
